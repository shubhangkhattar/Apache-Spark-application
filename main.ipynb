# Here, we have implemented queries mentioned in Assignment 4

  In the first cell, we have made an import of spark and connecting with local spark session to run our queries

```python
from pyspark.sql import SparkSession
import statistics
from statistics import mode
from pyspark.sql.types import DoubleType
import re
from pyspark.sql.functions import col

spark = SparkSession.builder.master("local[1]") \
                    .appName('Assignment4') \
                    .getOrCreate()
```

```python
sc = spark.sparkContext
```

###### Reading necessary data files by using appropriate function

```python
movies = spark.read.csv('ml-latest-small/ml-latest-small/movies.csv', header=True)
ratings = spark.read.csv('ml-latest-small/ml-latest-small/ratings.csv', header=True)
ratings = ratings.withColumn("rating",col("rating").cast(DoubleType()))
tags = spark.read.csv('ml-latest-small/ml-latest-small/tags.csv', header=True)

movies = movies.rdd
ratings = ratings.rdd
tags = tags.rdd
```

### From here, we are starting main section of Assignment 

#### Query 1

#### How many movies of genre Drama are there?

##### We are filtering movies dataset as want movies which has Drama genre and  we are counting instance of it

```python

print("Number of movies with Drama genre:", len(movies.filter(lambda m: m['genres'].find("Drama")>=0).collect()))
```

### Question 2

### Question 4
##### Compute min, average, max rating per movie.

This is simple query where we are using pyspark function as min, max and avg to sum/count for averaging the values

```python
rating_movie_pair = ratings.map(lambda r:(r.movieId, r.rating)).groupByKey().mapValues(list).collect()
```

```python
for i in rating_movie_pair:
    print(i[0], "max:", max(i[1]), "min:", min(i[1]), "avg:", sum(i[1])/len(i[1]))
```

### Question 5

##### Output data-set containing users that have rated a movie but not tagged it.

```python
rating_user_movie_pair = ratings.map(lambda r:(r.userId, r.movieId)).groupByKey().mapValues(list).collect()
tags_user_movie_pair = tags.map(lambda t:(t.userId, t.movieId)).groupByKey().mapValues(list).collect()
```

```python
def getlist(s):
        temp = [h[1] for h in tags_user_movie_pair if h[0]==s]
        if not temp:
            temp = ['empty']
            return temp
        else:
            return list(dict.fromkeys(temp[0]))
```

```python
user = []
for k in rating_user_movie_pair:
    t = getlist(k[0])
    check = set(k[1])-set(t)
    if len(check)>0:
        user.append(k[0])
print("Users who have rated but not tagged a movie:", user)
```


### Question 6

##### Output data-set containing users that have rated AND tagged a movie.

```python
user = []
for k in rating_user_movie_pair:
    t = getlist(k[0])
    check = set(k[1])-set(t)
    if len(check)!=len(k[1]):
        user.append(k[0])
            
print("users ",user)
```

### Question 7

##### Output dataset showing the number of movies per Genre per Year (movies will be counted many times if it's associated with multiple genres

```python
pattern = '\d{4}'
movies_title_genres = movies.map(lambda m: (m.movieId,m.title,(m.genres).split("|"),re.findall(pattern,m.title))).collect()
```

```python
genres = movies.map(lambda x: (x.genres).split("|")).collect()
genres = [item for sublist in genres for item in sublist]
genres = list(set(genres))
```

```python
def getMovieList(c):
    temp = [h for h in movies_title_genres if c in h[2]]
    peryear = [[[y[1],y[3],y[2]] for y in temp if x[3]==y[3]] for x in temp]
    return peryear
```

```python
perGenres = [[x,getMovieList(x)] for x in genres]

for y in perGenres[0][1]:
    print("\n----------------- year : ",y[0][1], "  -----------------genres :",perGenres[0][0])
    print("\n",y)
```
